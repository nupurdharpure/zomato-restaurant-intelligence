```python
data = pd.read_csv('zomato.csv', encoding='latin1')
# Loads your CSV file into a DataFrame called df. 
# The encoding='latin-1' part is important — Zomato data has some special characters (like café, naïve) that the default encoding can't read. Without this you'll get an error.
```

```python
data.drop_duplicates(inplace = True)
# Removes any rows that are exact copies of another row.
```

```python
data.dropna(subset=['Cuisines'], inplace = True)
# Drops only the rows where the Cuisines column is empty (there were 9 of them). 
# subset=['Cuisines'] means "only drop rows where THIS specific column is missing" — not the whole dataframe.
```

```python
data[data['Country Code'] == 1]
# Filters to keep only rows where Country Code equals 1 (which is India). 
# The inner part df['Country Code'] == 1 creates a True/False list, 
# and wrapping it in df[...] keeps only the True rows.
```

```python
data['Primary Cuisine'] = data['Cuisines'].apply(lambda x: x.split(',')[0].strip())
```

```python
# Creates a new column called Primary Cuisine. Many restaurants list multiple cuisines like "North Indian, Chinese, Mughlai". Here's what happens step by step:

# .split(',') → splits by comma → ['North Indian', ' Chinese', ' Mughlai']
# [0] → takes only the first item → 'North Indian'
# .strip() → removes any extra spaces around it
# lambda x: just means "for each value x in this column, do this"
```

```python
data.drop(columns=['Address', 'Locality Verbose', 'Rating color', 
                 'Switch to order menu', 'Currency'], inplace=True)
# Removes columns we won't need for our analysis. 
# Address and Locality Verbose are too detailed, 
# Rating color is redundant (we already have the number), 
# Currency is all the same after India filter, Switch to order menu has no useful data.
```

```python
data.reset_index(drop=True, inplace = True)
# After filtering rows, the index numbers have gaps (e.g. 0, 1, 5, 9...). 
# This resets them back to clean 0, 1, 2, 3... 
# drop=True means don't save the old index as a column.
```

```python
# Separate unrated restaurants
data_rated = data[data['Aggregate rating'] > 0]
data_unrated = data[data['Aggregate rating'] == 0]
print(f"Rated restaurants: {len(data_rated)}")
print(f"Unrated restaurants: {len(data_unrated)}")
```

```python
# We're creating two separate DataFrames instead of deleting the 0.0 rows entirely.

# data_rated → for all rating-based analysis (charts, averages, comparisons)
# data_unrated → we can use this separately — "restaurants with no reviews yet" is itself an insight
# f"..." is an f-string — it lets you put variables directly inside a string using {}
```

```python
data_rated.to_csv('zomato_rated.csv', index = False)
data_unrated.to_csv('zomato_unrated.csv', index = False)
```

```python
sql_queries = """
-- Query 1: Top cities by restaurant count
SELECT City, COUNT(*) AS total_restaurants, ROUND(AVG(`Aggregate rating`), 2) AS avg_rating
FROM restaurants
GROUP BY City
ORDER BY total_restaurants DESC
LIMIT 10;

-- Query 2: Most popular cuisines in Delhi
SELECT `Primary Cuisine`, COUNT(*) AS total_restaurants,
ROUND(AVG(`Aggregate rating`), 2) AS avg_rating,
ROUND(AVG(`Average Cost for two`), 0) AS avg_cost
FROM restaurants
WHERE City = 'New Delhi'
GROUP BY `Primary Cuisine`
ORDER BY total_restaurants DESC
LIMIT 10;

-- Query 3: Price range vs rating
SELECT `Price range`, COUNT(*) AS total_restaurants,
ROUND(AVG(`Aggregate rating`), 2) AS avg_rating,
ROUND(AVG(`Average Cost for two`), 0) AS avg_cost
FROM restaurants
GROUP BY `Price range`
ORDER BY `Price range` ASC;

-- Query 4: Online delivery analysis
SELECT `Has Online delivery`, COUNT(*) AS total_restaurants,
ROUND(AVG(`Aggregate rating`), 2) AS avg_rating,
ROUND(AVG(`Average Cost for two`), 0) AS avg_cost
FROM restaurants
GROUP BY `Has Online delivery`;

-- Query 5: Overrated restaurants in Delhi
SELECT `Restaurant Name`, City, `Primary Cuisine`,
`Aggregate rating`, Votes, `Average Cost for two`
FROM restaurants
WHERE Votes > 500 AND `Aggregate rating` < 3.5 AND City = 'New Delhi'
ORDER BY Votes DESC
LIMIT 10;
"""

with open('analysis.sql', 'w') as f:
    f.write(sql_queries)

print("analysis.sql saved!")
```
